```python
import sqlite3
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('data/Advertising.csv', index_col=0)
```

```python
engine = create_engine('sqlite:///data/advertising.db')
with engine.connect() as connection:
    df.to_sql('sales', con=connection, if_exists='append', index=False)
```

```python
connection = sqlite3.connect("data/advertising.db")
crsr = connection.cursor()
```

```python
select_all = "SELECT * FROM sales"
rows = crsr.execute(select_all).fetchall()

# Output to the console screen
for r in rows:
    print(r)

# Commiting the changes
connection.commit()
```

```python
def sql_query(query):

    # Ejecuta la query
    crsr.execute(query)

    # Almacena los datos de la query 
    ans = crsr.fetchall()

    # Obtenemos los nombres de las columnas de la tabla
    names = [description[0] for description in crsr.description]

    return pd.DataFrame(ans,columns=names)

query = '''
SELECT * 
FROM sales
'''
df = sql_query(query)
df
```
